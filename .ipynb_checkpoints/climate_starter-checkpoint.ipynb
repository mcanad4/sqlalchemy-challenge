# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect

# Imports the methods needed to abstract classes into tables
from sqlalchemy.ext.declarative import declarative_base

# Allow us to declare column types
from sqlalchemy import Column, Integer, String, Float 
```

```python
# Create an engine that can talk to the database
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# reflect the tables
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
# Assign the classes to a variables
Measurements = Base.classes.measurement
Stations = Base.classes.station
```

```python
# Create a session
session = Session(engine)
```

```python
# Create the inspector and connect it to the engine
inspector = inspect(engine)
```

```python
# Collect the names of tables within the database
inspector.get_table_names()
```

```python
# Using the inspector to print the column names within the 'Stations' table and its types
columns = inspector.get_columns('measurement')
for column in columns:
    print(column["name"], column["type"])
```

```python
# Using the inspector to print the column names within the 'Stations' table and its types
columns = inspector.get_columns('station')
for column in columns:
    print(column["name"], column["type"])
```

```python
engine.execute('SELECT * FROM station LIMIT 5').fetchall()
```

```python
engine.execute('SELECT * FROM measurement LIMIT 5').fetchall()
```

# Exploratory Precipitation Analysis

```python
# Total dates
session.query(func.count(Measurements.date)).all()
```

```python
# Earliest Date
session.query(Measurements.date).order_by(Measurements.date).first()
```

```python
# Latest Date
latest_date = session.query(Measurements.date).order_by(Measurements.date.desc()).first()
print(latest_date)
```

```python
import datetime as dt
```

```python
# Calculate the date one year from the last date in data set.
year_ago = dt.datetime(2017, 8, 23) - dt.timedelta(days=365)
print(year_ago)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
# Perform a query to retrieve the data and precipitation scores
session.query(Measurements.date,Measurements.prcp).\
    filter(Measurements.date > '2016-08-23').\
    order_by(Measurements.date).all()
```

```python
# Query for the date and precip amount
# for last year
# Sort the result by date
sel = [Measurements.date, 
       func.sum(Measurements.prcp)]
prcp_sums = session.query(*sel).\
    filter(Measurements.date > '2016-08-23').\
    group_by(Measurements.date).\
    order_by(Measurements.date).all()
prcp_sums
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
prcp_df = pd.DataFrame(prcp_sums, columns=['date', 'prcp'])

# Sort the dataframe by date
prcp_df = prcp_df.sort_values(by='date', ascending=True)
prcp_df.head()
```

```python
prcp_df.set_index('date', inplace=True)
prcp_df.plot.bar()
plt.ylabel('inches')
plt.tight_layout()
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
average = prcp_df["prcp"].mean()
average
```

```python
max = prcp_df["prcp"].max()
max
```

```python
min = prcp_df["prcp"].min()
min
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
sel = [Stations.station,
       Stations.name,
       Stations.latitude,
       Stations.longitude,
       Stations.elevation]
stations_sums = session.query(*sel)
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
station_df = pd.DataFrame(stations_sums, columns=['station','name','latitude','longitude','elevation'])
station_df.head()
```

```python
# Design a query to calculate the total number stations in the dataset
station_count = station_df["station"].count()
station_count
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
sel = [Measurements.station,
       func.count(Measurements.prcp)]
prcp_sums = session.query(*sel).\
    group_by(Measurements.station).all()
prcp_sums
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
sel = [Measurements.station, 
       func.max(Measurements.tobs), 
       func.min(Measurements.tobs), 
       func.avg(Measurements.tobs)]
station_stats = session.query(*sel).\
    filter(Measurements.station == 'USC00519281').\
    group_by(Measurements.station).all()
station_stats
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station
sel = [Measurements.station, Measurements.date, Measurements.tobs]
station_stats = session.query(*sel).\
    filter(Measurements.station == 'USC00519281').\
    filter(Measurements.date > '2016-08-23').\
    group_by(Measurements.station).all()
station_stats
```

```python
# Design a query to calculate the total number stations in the dataset
measure_sel = [Measurements.station,
       Measurements.date,
       Measurements.tobs,
       Measurements.prcp]
stations = session.query(*measure_sel)
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
measure_df = pd.DataFrame(stations, columns=['station','date','prcp','tobs'])
measure_df.head()
```

```python
# filter the dataframe for selected station and prior 12 months
measure_df = measure_df.loc[measure_df['station'] == 'USC00519281']
measure_df = measure_df.loc[measure_df['date'] >= '2016-08-23']
measure_df
```

```python
# Plot the results as a histogram
plt.hist(measure_df['prcp'])
plt.xlabel('Precipation')
plt.ylabel('Counts')
plt.show()
```
